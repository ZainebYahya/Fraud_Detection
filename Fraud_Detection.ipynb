```python
# will use  datetime.timedelta() to convert seconds to hours and minutes so it will be easier to read
time_delta = pd.to_timedelta(credit_card['Time'], unit= 's')
credit_card['minuts'] = (time_delta.dt.components.minutes).astype(int)
credit_card['hours'] = (time_delta.dt.components.hours).astype(int)
```

```python
sns.countplot(credit_card['Class']).set_xticklabels(['Not Fraud', 'Fraud'])
```

```python
# we can clearly see that the data is highly unbalanced we will use RandomUnderSampling
from imblearn.under_sampling import RandomUnderSampler 
under_sample = RandomUnderSampler(sampling_strategy= 0.5)
```

```python
X = credit_card.copy()
X = X.drop(['Class'],1)
```

```python
y = credit_card['Class'].values.reshape(-1,1)
```

```python
y.shape
```

```python
X.shape
```

```python
X_under , y_under = under_sample.fit_resample(X,y)
```

```python
Y = pd.DataFrame(y_under, columns= ['Class'])
```

```python
sns.countplot(Y['Class']).set_xticklabels(['Not Fraud', 'Fraud'])
```

```python
# big improvment as we see above 
# now split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_under,y_under, test_size = 0.2, random_state = 1)
```

```python
# now let's predict credit cards frauds with Logistic Regression
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(random_state=2)
logit  = model1.fit(X_train,y_train)
y_predict_logit = model1.predict(X_test)
```

```python
# let's check the scores
from sklearn import metrics
accuracy_logit = metrics.accuracy_score(y_test, y_predict_logit)
accuracy_logit
```

```python
preciesion_logit = metrics.precision_score(y_test, y_predict_logit)
preciesion_logit
```

```python
recall_logit = metrics.recall_score(y_test, y_predict_logit)
recall_logit
```

```python
f1_logit = metrics.f1_score(y_test, y_predict_logit)
f1_logit
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict_logit))
```

```python
# let's try random forest 
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
rf = model2.fit(X_train, y_train)
y_predict_rf= model2.predict(X_test)
```

```python
accuracy_rf = metrics.accuracy_score(y_test, y_predict_rf)
accuracy_rf
```

```python
preciesion_rf = metrics.precision_score(y_test, y_predict_rf)
preciesion_rf
```

```python
recall_rf = metrics.recall_score(y_test, y_predict_rf)
recall_rf
```

```python
f1_rf = metrics.f1_score(y_test, y_predict_rf)
f1_rf
```

```python
print(classification_report(y_test, y_predict_rf))
```
